# Reformating the orginal dataset

```python
import pandas as pd
```

```python
ratings = pd.read_csv("ratings.csv")
ratings.shape
```

```python
ratings.head()
```

```python
ratings.isnull().sum()
```

```python
ratings.dropna(inplace=True)
```

```python
ratings.isnull().sum()
```

We want to have the user_id as the index and the food_id as the columns and then have, as the cells, the ratings of the user for a certain food.

```python
ratings_matrix = ratings.pivot_table(index='User_ID', columns='Food_ID',values='Rating').fillna(0)
ratings_matrix
```

```python
ratings_matrix.to_csv("ratings_matrix.csv", index=False)
```
